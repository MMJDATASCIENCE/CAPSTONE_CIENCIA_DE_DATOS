```python

df = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788')
```

```python
df_code = df[0]
df_code.head(12)
```

```python
#Para depurar los datos
df_code2 = df_code[df_code['Borough']!= 'Not assigned']
df_code2 = df_code2.reset_index(drop=True)
df_code2.head(12)
```

```python
# Para saber si hay vecindarios no asignados
pregunta = df_code2[df_code2['Neighbourhood'] == 'Not assigned']
respuesta = pregunta.empty
if (respuesta):
    print ("No hay vecindarios asignados")

```

```python
# Determinamos el tamaño de dataframe.
print('El dataframe tiene {} filas'.format(df_code2.shape[0]), 'y {} columnas'.format(df_code2.shape[1]))
# Guardamos el dataframe como un archivo csv
df_code2.to_csv('df.csv', index=True)
```
