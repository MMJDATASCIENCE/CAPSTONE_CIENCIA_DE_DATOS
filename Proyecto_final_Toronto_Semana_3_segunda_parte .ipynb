```python
# Importación del paquete de geopy para determinar las coordenadas la ciudad de Toronto
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
# Buscamos las coordenadas para realizar el mapa
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Las coordenadas de '+address,' son {}, {}.'.format(latitude,longitude))
```

```python
#Generación de un mapa con los barrios superpuestos de Toronto
# Importamos el paquete Folium para realizar el mapa
!pip install folium
import folium
```

```python
# Gráfico del mapa de la ciudad de Toronto
map_toronto = folium.Map(location=[latitude, longitude], zoom_start = 10)
map_toronto
```

```python
# Añadimos marcadores de los municipios y vecindarios
for lat,lng, borough, neighborhood in zip(df_toronto_final['Latitude'],df_toronto_final['Longitude'],df_toronto_final['Borough'],df_toronto_final['Neighbourhood']):
    label= 'Neighborhood {}, Borough {}'.format(neighborhood,borough)
    label= folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup= label,
        color= 'red',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

map_toronto
```

```python
# Numero de vecindarios por municipios
df_toronto_borough = pd.value_counts(df_toronto_final['Borough'])
df_toronto_borough = pd.DataFrame(df_toronto_borough).reset_index()
df_toronto_borough = df_toronto_borough.rename(columns={'index':'Ciudad','Borough':'Número de vecindarios'})
df_toronto_borough
```
